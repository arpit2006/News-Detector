```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
fake = pd.read_csv("fake.csv")
true = pd.read_csv("true.csv")
```

```python
fake
```

```python
true
```

```python
fake["label"] = 0
true["label"] = 1
```

```python
#Concat Dataset
data = pd.concat([fake,true])
```

```python
#Shuffle data for proper predictions
data = data.sample(frac = 1,random_state = 42)
```

```python
#Initialize variables

#Input variable
x = data["text"]

#Output Variable
y = data["label"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
print("Training Shape: ",x_train.shape)
print("Training Shape: ",y_train.shape)
print("Testing Shape: ",y_test.shape)
print("Testing Shape: ",x_test.shape)
```

```python
#Use vectoriser to convert text to number
vectorizer = TfidfVectorizer(
    stop_words="english",
    max_df=0.7
)
```

```python
x_train_prepared = vectorizer.fit_transform(x_train)
x_test_prepared = vectorizer.transform(x_test)
```

```python
print("Training Shape: ",x_train_prepared.shape)
print("Testing Shape: ",x_test_prepared.shape)
```

```python
model = LogisticRegression()
model.fit(x_train_prepared,y_train)
lo_predictions = model.predict(x_test_prepared)
```

```python
accuracy = accuracy_score(y_test, lo_predictions)
print("Accuracy:", accuracy)
```

```python
print(classification_report(y_test, lo_predictions))
```
